```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
sns.lineplot(x=df['Hours'], y=df['Scores'])
```

```python
X = df[['Hours']]
y = df['Scores']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
pred_hour = np.array(6.5).reshape(-1,1)
```

```python
lr.predict(pred_hour)
```
